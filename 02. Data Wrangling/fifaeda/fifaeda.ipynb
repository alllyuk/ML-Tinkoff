# Есть ли зависмость потенциала (Potential) игрока от возраста? Всегда ли молодые игроки обладают большим потенциалом? 

```python
dfAgeMean = df.groupby('Age', as_index = False)['Potential'].mean()
dfAgeMax = df.groupby('Age', as_index = False)['Potential'].max()
fig = go.Figure()
fig.add_trace(go.Scatter(x = dfAgeMean['Age'], y = dfAgeMean['Potential'], name = 'Average Potential'))
fig.add_trace(go.Scatter(x = dfAgeMax['Age'], y = dfAgeMax['Potential'], name = 'Max Potential'))
fig.update_layout(title="Dependence of Potential from Age",
                  xaxis_title="Age",
                  yaxis_title="Potential",
                  legend = dict(y=0.97, x=0.7, xanchor="left", orientation="h"))
fig.show(renderer = 'colab')
```

*Как видно из графика, с возрастом средний потенциал плавно уменьшается и падает после 40 лет (будем считать высокий потенциал в 45 лет выбросом).*\
*Однако молодые игроки не обязательно обладают более высоким потенциалом, что видно из графика максимального его значения*

# Есть ли зависимость рейтинга игрока (Overall) от возраста? Когда у игроков "пик" карьеры? Игроков какого возраста больше всего с рейтингом 90+?  

```python
dfOverall = df.groupby('Age', as_index = False)['Overall'].median()
fig = go.Figure()
fig.add_trace(go.Scatter(x = dfOverall['Age'], y = dfOverall['Overall']))
fig.update_layout(title="Dependence of Overall from Age",
                  xaxis_title="Age",
                  yaxis_title="Overall",
                  legend = dict(y=0.97, x=0.7, xanchor="left", orientation="h"))
fig.show(renderer = 'colab')
```

*Игроки достигают высокого рейтинга приблизительно к 30 годам, зачем примерно до 40 лет он сохраняется на примерно одинаковом уровне, затем идет спад.*

```python
dfMaxOverall = df[df['Overall'] >= 90]
px.histogram(dfMaxOverall, x='Age', histfunc = 'count', nbins = 12, width = 500, height = 400, title = 'How many players with Overall 90+').show(renderer = 'colab')
```

*Таким образом, больше всего игроков с рейтингов 90+ в 32 года (четверо)*
